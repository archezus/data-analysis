```python
#SCRAPE NBA STATS FIRST PAGE
#https://www.datacamp.com/community/tutorials/web-scraping-using-python
from selenium.webdriver import Chrome
import pandas as pd
from bs4 import BeautifulSoup

#Download chromedriver from below link
#https://chromedriver.storage.googleapis.com/78.0.3904.70/chromedriver_win32.zip
#Relative location of the chromedriver
webdriver = "./ChromeDriver/chromedriver"

#This will open a Chrome browser controlled by the chromedriver
driver = Chrome(webdriver)
```

```python
#Our target for scraping
url = "https://stats.nba.com/leaders/"

driver.get(url)
```

```python
#Register the XPATH of next button
#You get this xpath from inspect element
python_button = driver.find_element_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[3]/div/div/a[2]')
```

```python
#Get column name
mytables = driver.find_elements_by_tag_name('table')
colname=mytables[0].text.split("\n")[0].split(" ")

#Create Initial DataFrame With Columns Name
df= pd.DataFrame(columns=colname)
df
```

```python
#Initialize
pnumb = []
pname = []
pstat = []
for page in range(6):
    if page!=0:
        #skip clicking page = 0
        python_button.click()
        
    mytables = driver.find_elements_by_tag_name('table')
    tabledata = mytables[0].text.split("\n")

    #Start from row-1 (row-0 contains columns name)
    for elem in tabledata[1:]:
        #Get number from 1, 4, 7,...
        if len(elem)<=3:
            pnumb.append(elem)
        #Get player name from 2, 5, 8,...
        elif len(elem)<=30:
            pname.append(elem)
        #Get stats from 3, 6, 9,..
        else:
            pstat.append(elem)
#Fill the column # with player number            
df["#"]= pnumb
#Fill the column PLAYER with player's name
df["PLAYER"] = pname
#Fill all the stats columns starting from col-2
i=0
for elem in pstat:
    df.loc[i][2:]=elem.split(" ")
    i = i + 1

#DON'T forget to close the chromedriver browser
driver.close()
```
